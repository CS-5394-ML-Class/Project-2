<b>Pre-merge Cleaning</b>

Before merging the data, we want to create two new columns. These columsn will sum up the hospital patients and ICU patients for later use.

```python
data_orig["hosp_icu_patients"] = data_orig["hosp_patients"] + data_orig["icu_patients"]
data_orig["hosp_icu_patients_per_million"] = data_orig["hosp_patients_per_million"] + data_orig["icu_patients_per_million"]
```

A quick check will be looking at rows with ISO codes of AFG. We know that the 2021 population count is 40,374,668
